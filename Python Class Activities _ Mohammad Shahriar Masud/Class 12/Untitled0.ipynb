```python
def celcius_to_fareinheit(celcius):
  conversion=lambda c:(9/4)*c+32
  return list(map(conversion,celcius))

Celcius_temp=[10,20,30,40,50]
print(celcius_to_fareinheit(Celcius_temp))
```

```python
def print_elements_with_index(numbers):
  for index,element in enumerate(numbers):
    print(f"Index {index} : {element}")

numbers=[1,2,3,4,5]
print_elements_with_index(numbers)
```

```python
def find_leargest_numbers(numbers):
  if not numbers:
    return None
  largest=numbers[0]
  for number in numbers:
    if number>largest:
      largest=number
  return largest
numbers=[1,2,3,4,5,6,78,90]
print(find_leargest_numbers(numbers))
```

```python
def analyze_numbers(numbers):
  minimum=min(numbers)
  maximum=max(numbers)
  sum_=sum(numbers)
  average=sum_/len(numbers)

  print(f"Minimum: {minimum}")
  print(f"Maximum: {maximum}")
  print(f"Sum: {sum_}")
  print(f"Average: {average}")

numbers=[]
for i in range(5):
  number=float(input(f"Enter number {i3}: "))
  numbers.append(number)

analyze_numbers(numbers)
```
