```python
import csv

def FindCourseAndStudentInfo(reader):
  CourseAndStudentInfo=dict()

  for row in reader:
    if row['Title'] in CourseAndStudentInfo:
      CourseAndStudentInfo[row['Title']].append(row['Student Id'])
    else:
      CourseAndStudentInfo[row['Title']]=[row['Student Id']]


  print(CourseAndStudentInfo)

try:
    with open('/content/SampleFile.csv', 'r') as file:
         reader = csv.DictReader(file)
         FindCourseAndStudentInfo(reader)
except:
    print("File does not exist")
finally:
    file.close()
```

```python
import csv

def FindCourseAndStudentInfo(reader):
  CourseAndStudentInfo=dict()

  for row in reader:
    if row['Student Name'] in CourseAndStudentInfo:
      CourseAndStudentInfo[row['Student Name']].append(row['Title'])
    else:
      CourseAndStudentInfo[row['Student Name']]=[row['Title']]


  print(CourseAndStudentInfo)

try:
    with open('/content/SampleFile.csv', 'r') as file:
         reader = csv.DictReader(file)
         FindCourseAndStudentInfo(reader)
except:
    print("File does not exist")
finally:
    file.close()
```
