```python
#Question no 5
temperatures = [28, 32, 35, 31, 29, 30, 33, 36, 27, 25, 34, 30, 29, 37, 38, 26, 31, 35, 33, 32, 36, 34, 29, 28, 27, 35, 32, 30, 31, 33]


above_32 = [temp for temp in temperatures if temp > 32]


classification = [(temp, "Hot" if temp > 35 else "Warm" if temp >= 30 else "Cool") for temp in temperatures]


average_temp = sum(temperatures) / len(temperatures)
temp_differences = [temp - average_temp for temp in temperatures]


print("Temperatures above 32°C:", above_32)
print("Temperature classifications:", classification)
print("Temperature differences from average:", temp_differences)
```

```python
#Question_no_4
import numpy as np

data = np.random.rand(10000)

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

min_value = np.min(data)
max_value = np.max(data)
count_greater_than_mean = np.sum(data > mean)

normalized_data = (data - min_value) / (max_value - min_value)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
print(f"Count of values greater than the mean: {count_greater_than_mean}")
print(f"First 10 normalized values: {normalized_data[:10]}")
```

```python
#question no 3
import numpy as np

A = np.array([[2, 4, 6], [1, 3, 5], [7, 9, 11]])
B = np.array([[8, 5, 2], [7, 4, 1], [6, 3, 0]])

sum_ab = A + B
product_ab = A * B
dot_product = np.dot(A, B.T)
det_A = np.linalg.det(A)

print("Element wise sum of A and B:")
print(sum_ab)

print("\nElementwise product of A and B:")
print(product_ab)

print("\nDot product of A and B^T:")
print(dot_product)

print("\nDet of A:")
print(det_A)
```
