```python
list_of_word = ["Isaac Newton (1642-1727) was an English physicist, mathematician, astronomer, theologian, and writer who is considered one of the most influential scientists of all time: "]
wcounts = dict()
for line in list_of_word:
    for w in line.split():
        if w in wcounts:
            wcounts[w] += 1
        else:
            wcounts[w] = 1
print(wcounts)
```
