```python
#Python Programto get 5 words from the user iterate that list to display only those word that starts with"b"
word_list = []
for i in range(5):
    word = input(f"Enter word {i+1}: ")
    word_list.append(word)
for word in word_list:
    if word.startswith("b"):
        print(word)
```
