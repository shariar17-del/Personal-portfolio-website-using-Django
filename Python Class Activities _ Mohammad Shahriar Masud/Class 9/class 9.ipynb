```python
a=int(input("enter the first number"))
b=int(input("enter the second number"))
product_of_two_numbers=a*b
print(f'the product of {a} and {b} is {product_of_two_numbers}')
sum_of_two_numbers=a+b
print(f'the sum of {a} and {b} is {sum_of_two_numbers}')
difference_of_two_numbers=a-b
print(f'the difference of {a} and {b} is {difference_of_two_numbers}')
quotient_of_two_numbers=a/b
print(f'the quotient of {a} and {b} is {int(quotient_of_two_numbers)}')
```
