```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
```

```python
array = np.array([1, 2, 3, 4])
print(np.sum(array))
print(np.mean(array))
print(np.std(array))
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(np.dot(a, b))
```

```python
a = np.array([1, 2, 3])
b = 2
print(a * b)
```

```python
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)
```

```python
a = np.array([1, 2, 3, 4])
b = a[2:4] # Creating a view of the array
b[0] = 100 # Modify the view
print(a)
```
